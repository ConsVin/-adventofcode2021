# Brute force

```python
import numpy as np
fname = 'input'
s = open(fname,'r').read().replace("| ","")
symbols = [line.split(" ") for line in s.split("\n")]
symbols_len = [[len(word) for word in line] for line in symbols]

values = [line[10:] for line in symbols_len]

values
```

```python
from pprint import pprint
digits = {  0 : 'abcefg',
            1 : 'cf',
            2 : 'acdeg',
            3 : 'acdfg',
            4 : 'bcdf',
            5 : 'abdfg',
            6 : 'abdefg',
            7 : 'acf',
            8 : 'abcdefg',
            9 : 'abcdfg'}
digits_len = { k:len(digits[k]) for k in digits }
uniq_digits = [1, 4, 7, 8]
uniq_lens =  [digits_len[u] for u in uniq_digits]
uniq_lens
```

```python
import numpy as np
x = [[ (sym in uniq_lens) for sym in row ] for row in values]
np.sum(x)

```

```python
digits = {  0 : 'abcefg',
            1 : 'cf',
            2 : 'acdeg',
            3 : 'acdfg',
            4 : 'bcdf',
            5 : 'abdfg',
            6 : 'abdefg',
            7 : 'acf',
            8 : 'abcdefg',
            9 : 'abcdfg'}
letter2int = { c:ord(c) - ord('a') for c in 'abcdefg'}
digits_list = [ [ letter2int[char] for char in digits[d]] for d in digits]
digits_list
mat = np.zeros((10,7),dtype='int8')
for i,d in enumerate(digits_list):
    for j in d:
        mat[i][j]=1
#print( np.unique(  np.sum(mat,axis=0)), return_counts=True  ) 
print(np.unique(  np.sum(mat,axis=0), return_counts=True  ))
print(np.unique(  np.sum(mat,axis=1), return_counts=True  ))

np.sum(mat,axis=0)
```
