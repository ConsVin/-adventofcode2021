```python
import numpy as np
fname = 'sample'
line_cmd = open(fname,'r').read()
print(line_cmd)
s = line_cmd.replace("=",",").replace("..",",").split(",")
x_from,x_to = int(s[1]),int(s[2])
y_from,y_to = int(s[4]),int(s[5])
print(f"{x_from}..{x_to} , y = {y_from}..{y_to}")

x_offset =   0;
y_offset =  10;
class Area():
    def __init__(self, x_low, x_up, y_low, y_up):
        x_size =  x_up - x_low
        y_size =  y_up - y_low
        self.x_low = x_low
        self.y_low = y_low
        self.area = np.zeros((y_size,x_size))
        self.set_val(0,0,9)
    def set_val(self, x_val, y_val, val):
        x_abs = x_val - self.x_low
        y_abs = y_val - self.y_low
        self.area[y_abs,x_abs] = val
    def get_val(self, x_val, y_val, val):
        x_abs = x_val - self.x_low
        y_abs = y_val - self.y_low
        return self.area[y_abs,x_abs]
    def set_space(self, x_from, x_to, y_from, y_to):
        for x in range(x_from,x_to+1):
            for y in range(y_from,y_to+1):
                self.set_val(x,y,1)
        
    def display(self):
        cast = {
            0:'.',
            1:'T',
            2:'#',
            9:'S'
        }
        for row in self.area[::-1]:
            line = ''.join([cast[c] for c in row])
            print(line)
        
A = Area(0,32,-12,+17)
A.set_space(x_from,x_to,y_from,y_to )

def x_change(x):
    if x>0:
        return -1
    elif x<0:
        return +1
    else:
        return 0
x_vel,y_vel = 7,5
x,y = 0,0

for _ in range(15):
    x+=x_vel
    y+=y_vel
    res = A.set_val(x,y,2)
    y_vel += -1
    x_vel += x_change(x_vel)
    print([x,y,y_vel,x_vel])
A.display()
```
