```python
import numpy as np
fname = 'sample'

lines = open(fname,'r').read().split("\n")
o2c = {
    '(':')',
    '[':']',
    '{': '}',
    '<': '>',
}
err_scores ={
    ')':3,
    ']':57,
    '}':1197,
    '>':25137
}
complete_score =  {
    ')':1,
    ']':2,
    '}':3,
     '>':4}

c2o ={ o2c[k]:k for k in o2c}
o = list(o2c.keys())
c = list(c2o.keys())

score_list = []
for l in lines:
    stack = []
    print(l)
    err = False
    for c in l:
        if c in o:
            stack.append(c)
        else:
            prev_openned = stack.pop()
            if prev_openned != c2o[c]:
                err = True
                break
    if not(err):
        print(f"Incomplete line : {l}, stack {''.join(stack)}")
        append_line = [ o2c[s] for s in stack]
        score= 0
        for a in reversed(append_line):
            score*=5
            score+=complete_score[a]
        print(score)
        score_list.append(score) 
print(f"Ans : {score_list}")
```
