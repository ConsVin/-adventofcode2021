```python
import numpy as np
import matplotlib.pyplot as plt

fname = 'input'

lines = open(fname,'r').read().split("\n")
map_x = np.array( [ [int(c) for c in row] for row in lines] )
map_pad = np.pad(map_x, pad_width=1,constant_values=10)
minima_list = []
minima_points = []

one_step_map = 0*map_pad;

for r in range(1,map_x.shape[0]+1):
    for c in range(1,map_x.shape[1]+1):
        cell = map_pad[r,c]
        adj = [    map_pad[r+1,c], 
                   map_pad[r-1,c],
                   map_pad[r,c+1],
                   map_pad[r,c-1]]
        if np.all(cell<adj):
            minima_list.append(cell)
            minima_points.append([r,c])
            one_step_map[r,c]=1
ans = np.sum(np.array(minima_list)+1)
print(f"{minima_list} => \n {ans}")

N_LOOPS = 20
for n_loop in range(N_LOOPS):
    useless_loop = True
    for r in range(1,map_x.shape[0]+1):
        for c in range(1,map_x.shape[1]+1):
            if (one_step_map[r,c] == 1):
                cell = map_pad[r,c]
                for pr,pc in [ [r+1, c], [r-1, c], [r, c+1], [r,c-1]]:
                    n = map_pad[pr,pc]
                    if ((n <9) and (n>cell)):
                        one_step_map[pr,pc] = 1
cluster_cnt = 0
cluster_map = 0*one_step_map
cluster_cnt = 0
cluster_size = 0

def grow_cluster(r,c):
    if (one_step_map[r,c]==0) or (cluster_map[r,c]>0):
        return
    cluster_map[r,c] = cluster_cnt
    global cluster_size 
    cluster_size +=1
    for pr,pc in [ [r+1, c], [r-1, c], [r, c+1], [r,c-1]]:
        grow_cluster(pr,pc);
clusters_size_list = []
for r in range(1,map_x.shape[0]+1):
    for c in range(1,map_x.shape[1]+1):
        if (one_step_map[r,c]==1) and (cluster_map[r,c]==0):
            cluster_cnt+=1;
            cluster_size =0
            grow_cluster(r,c)
            clusters_size_list.append(cluster_size)
            
                
plt.matshow(one_step_map)
plt.show()

plt.figure(figsize=(10,5))
plt.matshow(cluster_map,fignum=1)


top3 = np.sort(clusters_size_list)[-3:]
print(np.sort(clusters_size_list))
print(f"{top3} => {np.prod(top3)}")
```
