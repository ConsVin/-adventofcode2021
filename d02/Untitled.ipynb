# P1


```python
fname = 'input_a'
pos   = 0
depth = 0
with open(fname, 'r') as file:
    data = file.read().split("\n")
for d in data:
    cmd,arg = d.split(" ")
    arg=int(arg)
    if cmd=='forward':
        pos+=arg
    elif cmd=='down':
        depth+=arg
    elif cmd=='up':
        depth-=arg
    print([cmd,arg,depth, pos])        
print([depth,pos])
print(depth*pos)
    

```

# P2

```python
fname = 'input_a'
pos   =  0
depth =  0
aim   =  0
log_list = []
with open(fname, 'r') as file:
    data = file.read().split("\n")
for d in data:
    cmd,arg = d.split(" ")
    arg=int(arg)
    if cmd=='forward':
        pos   += arg
        depth += aim*arg
        
        
    elif cmd=='down':
        aim+=arg
    elif cmd=='up':
        aim-=arg
    print([cmd,arg,depth, pos])        
    log_list.append([depth, pos])
print([depth,pos])
print(depth*pos)
    

```

```python
import numpy as np
coord = np.array(log_list).T
import matplotlib.pyplot as plt
plt.plot(coord[0])
```
