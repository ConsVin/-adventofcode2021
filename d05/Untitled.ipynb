```python
import numpy as np
fname = 'input'
with open(fname,'r') as file:
    data = file.read()
x = np.fromstring(data.replace(" -> ",',').replace("\n",','),sep=',',dtype='int32')
lines = x.reshape(int(len(x)/4),4)

hor_or_vert = []
for line in lines:
    if line[0]==line[2] or line[1]==line[3]:
        hor_or_vert.append(line)
hor_or_vert
L = 1000
area = np.zeros((L,L)).astype(int)
print("----")
for l in hor_or_vert:
    x0    = min(l[1],l[3])
    x_len = abs(l[1] - l[3])
    y_len = abs(l[0] - l[2])
    y0    = min(l[0],l[2])
    
    area[x0:x0+x_len+1,y0:y0+y_len+1  ]+=1
np.sum(area>=2)
```

```python
import numpy as np
fname = 'input'
with open(fname,'r') as file:
    data = file.read()
x = np.fromstring(data.replace(" -> ",',').replace("\n",','),sep=',',dtype='int32')
lines = x.reshape(int(len(x)/4),4)

L = 1000
area = np.zeros((L,L)).astype(int)

def take_sign(x):
    if x==0:
        return 0
    elif x<0:
        return -1
    else:
        return 1
for l in lines:
    x0 = l[0]
    y0 = l[1]
    
    x_dir   = l[2] - l[0]
    y_dir   = l[3] - l[1]
    
    x_step = take_sign(x_dir)
    y_step = take_sign(y_dir)
    n_steps=max(abs(x_dir),abs(y_dir))    
    
    print(f"{x0},    {y0} ,   {x_step},  {y_step} {n_steps}")
    for i in range(n_steps+1):
        area[x0+x_step*i,y0+y_step*i]+=1
area
    
    
    
    
    
    
print(area)         
np.sum(area>=2)
```
